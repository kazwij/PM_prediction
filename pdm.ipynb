```python
df.columns
```

```python
#re-order the column
df = df[['Air temperature', 'Process temperature', 'Rotational speed', 'Torque',
       'Tool wear','Power' ,'Machine failure']]


df.head()
```

```python
# X, y seperation

X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

## Modeling and Evaluating  ##

#### over sample  ##

```python
# importing SMOTE, SVMSMOTE,RandomOverSampler from imbalance library
from imblearn.over_sampling import SMOTE, SVMSMOTE,RandomOverSampler
# using RandomOverSampler
oversamp = RandomOverSampler(random_state=0)
```

```python
# splitting the dataset ensuring same distrubution of target variable 'y'
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 0,
                                                    stratify=y)
```

```python
# data scaling 
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# importing model evaluation criterias 
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score,roc_auc_score, ConfusionMatrixDisplay,matthews_corrcoef
from sklearn.metrics import ConfusionMatrixDisplay  
```

```python
#model check list

model_performance = pd.DataFrame(columns=['Accuracy','Recall','Precision','F1-Score','MCC score','time to train','time to predict','total time'])
```

```python
#  model import 
def model(name):
    print(name)
    start = time.time()
    model = name().fit(X_train,y_train)
    end_train = time.time()
    y_predictions = model.predict(X_test) # These are the predictions from the test data.
    end_predict = time.time()
    training_time = end_train - start
    pred_time   = end_predict - end_train
    #print(f"Time taken for training: {round((end_train - start) * 1000,0)} ms")
    #print(f"Time taken for prediction: {round((end_predict - end_train) * 1000,0)} ms")
    #print(f"Total time taken for the model :{round((training_time+pred_time)*1000,0)} ms")
    
    
    accuracy = accuracy_score(y_test, y_predictions)
    recall = recall_score(y_test, y_predictions, average='weighted')
    precision = precision_score(y_test, y_predictions, average='weighted')
    f1s = f1_score(y_test, y_predictions, average='weighted')
    MCC = matthews_corrcoef(y_test, y_predictions)
    # ROC_AUC = roc_auc_score(y_test, y_predictions, average='weighted')
    ROC_AUC = roc_auc_score(y_test, model.predict_proba(X_test)[:,1], average='weighted')
    
    #get the model name 
    model_name = type(model).__name__
    
    
    print('')
    print("Accuracy: "+ "{:.2%}".format(accuracy))
    print("Recall: "+ "{:.2%}".format(recall))
    print("Precision: "+ "{:.2%}".format(precision))
    print("F1-Score: "+ "{:.2%}".format(f1s))
    print("MCC: "+ "{:.2%}".format(MCC))
    print("ROC AUC score: "+ "{:.2%}".format(ROC_AUC))
    print("time to train: "+ "{:.2f}".format(end_train-start)+" s")
    print("time to predict: "+"{:.2f}".format(end_predict-end_train)+" s")
    print("total: "+"{:.2f}".format(end_predict-start)+" s")
    model_performance.loc[model_name] = [accuracy, recall, precision, f1s,MCC,end_train-start,end_predict-end_train,end_predict-start]
    
   
    # plot the confustion metrix 
    plt.rcParams['figure.figsize']=5,5 
    sns.set_style("white")
    ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, cmap=plt.cm.Blues)  
    plt.title(f'model {model_name} ')
    plt.show()
```
